```python
import numpy as np
import pandas as pd

# Parameters
days = 30  # Simulation period in days
prosumers = 10  # Number of prosumers
consumers = 50  # Number of consumers

# Simulate daily production and consumption
daily_production = np.random.normal(27, 3, (days, prosumers))  # kWh, production by prosumers
daily_consumption = np.random.normal(20, 5, (days, consumers))  # kWh, consumption by consumers

# Aggregate total production and consumption
total_production = daily_production.sum(axis=1)  # Total production per day
total_consumption = daily_consumption.sum(axis=1)  # Total consumption per day

# Calculate surplus energy
surplus_energy = total_production - total_consumption

# Pricing
p2p_rate = 0.18  # €/kWh, Peer-to-Peer rate
grid_rate = 0.15  # €/kWh, Grid rate

# Calculate revenues
p2p_sales = np.clip(surplus_energy, 0, None)  # Only surplus energy sold in P2P market
grid_sales = np.clip(surplus_energy - p2p_sales, 0, None)  # Remaining surplus sold to the grid

revenue_p2p = p2p_sales * p2p_rate
revenue_grid = grid_sales * grid_rate

# Create the simulation DataFrame
simulation_data = pd.DataFrame({
    'Day': np.arange(1, days + 1),
    'Total_Production (kWh)': total_production,
    'Total_Consumption (kWh)': total_consumption,
    'Surplus_Energy (kWh)': surplus_energy,
    'P2P_Sales (kWh)': p2p_sales,
    'Grid_Sales (kWh)': grid_sales,
    'Revenue_P2P (€)': revenue_p2p,
    'Revenue_Grid (€)': revenue_grid,
    'Total_Revenue (€)': revenue_p2p + revenue_grid
})

# Save the simulation data to a CSV file
output_path = "energy_simulation_results.csv"  # File name
simulation_data.to_csv(output_path, index=False)

print(f"Simulation data saved to {output_path}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Simulation parameters
days = 30
prosumers = 20
consumers = 50

# Simulate daily energy production and consumption
np.random.seed(42)  # For reproducibility
daily_production = np.random.normal(100, 10, (days, prosumers))  # kWh
daily_consumption = np.random.normal(25, 5, (days, consumers))   # kWh

# Calculate total daily production and consumption
total_production = daily_production.sum(axis=1)
total_consumption = daily_consumption.sum(axis=1)

# Calculate surplus energy
surplus_energy = total_production - total_consumption

# Pricing rates
p2p_rate = 0.18  # €/kWh
grid_rate = 0.12  # €/kWh

# Calculate potential revenues
p2p_sales = np.clip(surplus_energy, 0, None)
grid_sales = np.clip(surplus_energy - p2p_sales, 0, None)
revenue_p2p = p2p_sales * p2p_rate
revenue_grid = grid_sales * grid_rate
total_revenue = revenue_p2p + revenue_grid

# Create a DataFrame to store the results
simulation_data = pd.DataFrame({
    'Day': np.arange(1, days + 1),
    'Total Production (kWh)': total_production,
    'Total Consumption (kWh)': total_consumption,
    'Surplus Energy (kWh)': surplus_energy,
    'P2P Sales (kWh)': p2p_sales,
    'Grid Sales (kWh)': grid_sales,
    'Revenue P2P (€)': revenue_p2p,
    'Revenue Grid (€)': revenue_grid,
    'Total Revenue (€)': total_revenue
})

# Plotting the results
plt.figure(figsize=(14, 8))

# Total Production and Consumption
plt.subplot(2, 1, 1)
plt.plot(simulation_data['Day'], simulation_data['Total Production (kWh)'], label='Total Production (kWh)', color='green')
plt.plot(simulation_data['Day'], simulation_data['Total Consumption (kWh)'], label='Total Consumption (kWh)', color='red')
plt.fill_between(simulation_data['Day'], simulation_data['Total Production (kWh)'], simulation_data['Total Consumption (kWh)'],
                 where=(simulation_data['Total Production (kWh)'] >= simulation_data['Total Consumption (kWh)']),
                 interpolate=True, color='green', alpha=0.3, label='Surplus Energy')
plt.xlabel('Day')
plt.ylabel('Energy (kWh)')
plt.title('Daily Total Energy Production and Consumption')
plt.legend()
plt.grid(True)

# Revenues from P2P and Grid Sales
plt.subplot(2, 1, 2)
plt.plot(simulation_data['Day'], simulation_data['Revenue P2P (€)'], label='Revenue P2P (€)', color='blue')
plt.plot(simulation_data['Day'], simulation_data['Revenue Grid (€)'], label='Revenue Grid (€)', color='orange')
plt.plot(simulation_data['Day'], simulation_data['Total Revenue (€)'], label='Total Revenue (€)', color='purple', linestyle='--')
plt.xlabel('Day')
plt.ylabel('Revenue (€)')
plt.title('Daily Revenue from P2P and Grid Sales')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Display summary statistics
summary_stats = simulation_data.describe().transpose()
print(summary_stats)
```
